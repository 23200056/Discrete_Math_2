\begin{equation}

\end{equation}

\begin{equation}
E[Y_i] = \sum_{y=0}^{n} y \cdot \binom{n}{y} p_i^y (1-p_i)^{n-y}
\end{equation}

\begin{equation}
E[Y_i] = n \cdot p_i
\end{equation}

##### QUESTION 1:

For $( Y = k )$ (where $( k )$ successes occur in $( n )$ trials), there are $(\binom{n}{k})$ ways to choose which $( k )$ trials result in success out of $( n )$ trials.

The probability of any specific sequence of $( k )$ successes and $( n - k )$ failures is $( p^k q^{n-k} )$, given the independence of trials.

The probability of observing exactly $( k )$ successes in $( n )$ trials is the number of such sequences multiplied by the probability of any one sequence. Therefore,

\begin{equation}
P(Y = k) = \binom{n}{k} p^k q^{n-k}
\end{equation}

Since $q = 1 - p$, we can substitute this in the equation.

\begin{equation}
P(Y = k) = \binom{n}{k} p^k (1 - p)^{n-k}

##### QUESTION 2

To show that the expectation of the binomial distribution is $( E[Y] = np )$, we start with the definition of expectation for a discrete random variable. For a binomially distributed random variable $( Y \sim \text{Binomial}(n, p) )$, the expectation $( E[Y] )$ is given by:

\begin{equation}
E[Y] = \sum_{k=0}^{n} k \cdot P(Y = k)
\end{equation}

From the binomial distribution, we know that:

\begin{equation}
P(Y = k) = \binom{n}{k} p^k (1 - p)^{n - k}
\end{equation}

Substituting this into the expectation, we have:

\begin{equation}
E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1 - p)^{n - k}
\end{equation}

To simplify this, we use the properties of binomial coefficients and algebraic manipulation. Note that:

\begin{equation}
k \cdot \binom{n}{k} = k \cdot \frac{n!}{k!(n-k)!} = \frac{n!}{(k-1)!(n-k)!} = n \cdot \frac{(n-1)!}{(k-1)!(n-k)!} = n \cdot \binom{n-1}{k-1}
\end{equation}

Thus,

\begin{equation}
E[Y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1 - p)^{n - k} = \sum_{k=1}^{n} n \cdot \binom{n-1}{k-1} p^k (1 - p)^{n - k}
\end{equation}

Note that for $( k = 0 )$, the term is zero because $( k \cdot P(Y = 0) = 0 )$. Therefore, we can adjust the sum to start from $( k = 1 )$.

We can factor out $( n \cdot p )$ from the sum:

\begin{equation}
E[Y] = n \cdot p \sum_{k=1}^{n} \binom{n-1}{k-1} p^{k-1} (1 - p)^{(n-1)-(k-1)}
\end{equation}

Notice that the summation now represents the binomial expansion of $( (p + (1 - p))^{n-1} )$, which equals $( 1 )$:

\begin{equation}
\sum_{k=1}^{n-1} \binom{n-1}{k-1} p^{k-1} (1 - p)^{(n-1)-(k-1)} = (p + (1 - p))^{n-1} = 1
\end{equation}

Thus,

\begin{equation}
E[Y] = n \cdot p \cdot 1 = np
\end{equation}

Therefore, we have shown that the expected value of a binomially distributed random variable $( Y )$ is:

\begin{equation}
E[Y] = np
\end{equation}

### Task 3: 
Consider a probility distribution of $p(y) = \frac{\lambda^y}{y!} e^{-\lambda}$.

Question 1: Find a general formula for the expected value of this distribution.

Question 2: A certain type of tree has seedlings randomly dispersed in a large area, with the mean density of seedlings being approximately five per square yard. If a forester randomly locates ten 1-square-yard sampling regions in the area, find the probability that none of the regions will contain seedlings.


#### ANSWER:

##### QUESTION 1:

The given probability distribution is a Poisson distribution. The expected value (mean) of a Poisson distribution is equal to its parameter, λ. So, the expected value of this distribution is $( \lambda )$.

The expected value E(Y) of a random variable Y is defined as:

\begin{equation}
E(Y) = \sum_{y=0}^{\infty} y \cdot p(y)
\end{equation}

For a Poisson distribution, this becomes:

\begin{equation}
E(Y) = \sum_{y=0}^{\infty} y \cdot \frac{\lambda^y}{y!} e^{-\lambda}
\end{equation}

This can be simplified to:

\begin{equation}
E(Y) = \lambda e^{-\lambda} \sum_{y=1}^{\infty} \frac{\lambda^{y-1}}{(y-1)!}
\end{equation}

The sum in this expression is the sum of probabilities for a Poisson distribution with parameter λ, which sums to 1. Therefore, the expected value is:

\begin{equation}
E(Y) = \lambda
\end{equation}

##### QUESTION 2:

The number of seedlings in a square yard follows a Poisson distribution with λ = 5. We want to find the probability that none of the 10 regions contain seedlings. 

The probability that a single region contains no seedlings is given by the formula for the Poisson distribution with y = 0:

\begin{equation}
p(0) = \frac{\lambda^0}{0!} e^{-\lambda} = e^{-\lambda}
\end{equation}

Substituting λ = 5, we get:

\begin{equation}
p(0) = e^{-5}$$
\end{equation}

Since the regions are independent, the probability that all 10 regions contain no seedlings is the product of their individual probabilities:

\begin{equation}
p(\text{all empty}) = (e^{-5})^{10} = e^{-50}
\end{equation}

So, the probability that none of the 10 regions will contain seedlings is $( e^{-50} )$.
