### Graph

A graph is a structure consisting of vertices (nodes) and edges connecting pairs of vertices. Graphs can be directed or undirected and are used to model relationships and interactions in various fields.

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_node('F')
G.add_edges_from([('X','Z'),('Y','X'),('Z','Y')])

pos = nx.spring_layout(G)

nx.draw_networkx_nodes(G,pos, node_size=500)
nx.draw_networkx_edges(G,pos)
nx.draw_networkx_labels(G,pos)
plt.show()
```

# Directed Graphs

A directed graph, is a type of graph where edges have a direction, indicated by arrows pointing from one vertex to another. This structure is used to represent asymmetric relationships such as one-way streets or hierarchical structures.

```python
G = nx.DiGraph()
G.add_edges_from([('X','Z'),('Y','X'),('Z','Y')])

pos = nx.circular_layout(G)

nx.draw_networkx_nodes(G,pos, node_size=500)
nx.draw_networkx_edges(G,pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G,pos)
plt.show()
```

# Simple Graphs

A simple graph is an undirected graph where there is at most one edge between any pair of vertices, and no edges connect a vertex to itself. 

```python
G = nx.Graph()
G.add_edges_from([('X','Z'),('Y','X'),('Z','Y'), ('X', 'W')])

pos = nx.shell_layout(G)

nx.draw_networkx_nodes(G,pos, node_size=500)
nx.draw_networkx_edges(G,pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G,pos)
plt.show()
```

# Complete Graphs

A complete graph is a simple graph where each pair of distinct vertices is connected by a unique edge. In other words, every vertex is directly connected to every other vertex in the graph. Complete graphs are denoted by $K_n$ where $n$ is the number of vertices. They are useful in various mathematical and theoretical contexts, including graph theory and combinatorics.

```python
G = nx.Graph()
G.add_edges_from([('X','Y'),('X','Z'),('Y','Z'),('P','Q'), ('R', 'S')])  
G.add_node ('T')

pos = nx.kamada_kawai_layout(G)

nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G, pos)
plt.show()
```

# Bipartite Graphs

Bipartite graphs are graphs whose vertices can be divided into two disjoint sets such that no two vertices within the same set are adjacent. In other words, the vertices are partitioned into two groups, and all edges connect vertices from different groups. Bipartite graphs are often used to model relationships between two distinct types of entities.

```python
G = nx.Graph()
G.add_nodes_from([1,2,3], bipartite=0) 
G.add_nodes_from([4,5,6], bipartite=1)
G.add_edges_from([(1,4), (1,5), (1,6)
                ,(2,4), (2,5), (2,6), (3,4)])

top = nx.bipartite.sets(G)[0]
pos = nx.bipartite_layout(G, top)
nx.draw(G, pos=pos, with_labels=True, node_color=['cyan', 'magenta', 'yellow', 'red', 'green', 'blue'])
plt.show()
```

# Subgraphs


A subgraph of a graph $G$ is a graph formed by selecting a subset of vertices and edges from $G$ while preserving the original structure. In other words, it contains a subset of the vertices of the original graph along with any edges that connect those vertices. Subgraphs are useful for analyzing specific parts of a larger graph and for studying properties of graphs in more detail.

```python
G = nx.Graph()
H = nx.Graph()
G.add_edges_from([('X','Y'),('X','Z'),('Z','P'),('Y','P'),('X','P'),('Y','Z'), ('W', 'X')])
H.add_edges_from([('X','Y'),('Y','P'),('P','X'), ('W', 'X')])

pos = {'X': (0, 0), 'Y': (-1, 1), 'P': (0, 2), 'Z': (1, 1), 'W': (2, 2)}

plt.figure(figsize=(10, 5))  

plt.subplot(1, 2, 1)  
nx.draw_networkx_nodes(G, pos, node_size=500)
nx.draw_networkx_edges(G, pos, edgelist=G.edges(), edge_color='black')
nx.draw_networkx_labels(G, pos)
plt.title('Graph G')
plt.axis('off')  

plt.subplot(1, 2, 2)  
nx.draw_networkx_nodes(H, pos, node_size=500)
nx.draw_networkx_edges(H, pos, edgelist=H.edges(), edge_color='black')
nx.draw_networkx_labels(H, pos)
plt.title('Graph H')
plt.axis('off')  

plt.tight_layout()  
plt.show()
```

# Euler's Circuit

Euler's circuit is a circuit in a graph that traverses each edge exactly once and returns to the starting vertex. In other words, it is a closed walk that covers every edge of the graph exactly once. A graph has an Eulerian circuit if and only if every vertex has an even degree.

```python
G = nx.Graph()
for node in ['X', 'Y', 'Z', 'P', 'W']:
    G.add_node(node)

edges = [('X', 'Y'), ('Y', 'Z'), ('Z', 'P'), ('P', 'X'), ('W', 'X')]

for edge in edges:
    G.add_edge(*edge)

nx.draw(G,with_labels=True, node_color='deepskyblue', edge_color='gray')
plt.show()
```
