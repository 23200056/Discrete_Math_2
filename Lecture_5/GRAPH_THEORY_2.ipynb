# Matrix Representations of Graphs
---

- An $m × n$ (read “m by n” referred to as the dimension or size) matrix A over a set S is a rectangular array of elements of S arranged into m rows and n columns:
\begin{equation}
\textbf{A} = 
\begin{bmatrix}
a_{11} & a_{12} & ... & a_{1j} & ... & a_{1n} \\
a_{21} & a_{22} & ... & a_{2j} & ... & a_{2n} \\
. & . & ... & ... & ... \\
. & . & ... & ... & ... \\
. & . & ...  & ... & ... \\
a_{i1} & a_{i2} & ... & a_{ij} & ... & a_{in} \\
. & . & ... & ... & ... \\
. & . & ... & ... & ... \\
. & . & ...  & ... & ... \\
a_{m1} & a_{m2} & ... & a_{mj} & ... & a_{mn} \\
\end{bmatrix}
\end{equation}

- The matrix can also be re-written as: $\textbf{A} = \begin{bmatrix}a_{ij}\end{bmatrix}$.

- The $\textbf{i}^{\textbf{th}} \textbf{row}$ is $\begin{bmatrix} a_{i1} & a_{i2} & ... & a_{ij} & ... & a_{in}\end{bmatrix}$

- The $\textbf{j}^{\textbf{th}} \textbf{column}$ is $\begin{bmatrix} a_{1j} \\ a_{2j} \\ ... \\ a_{ij} \\ ... \\ a_{mj}\end{bmatrix}$

\end{equation}

# Matrix Representation of an Undirected Graphs.
---

$\textbf{Definition}$

- Let $G$ be an undirected graph with ordered vertices $v_1,v_2,...,v_n$. The adjacency matrix of $G$ is the $n × n$ matrix $\textbf{A} = \begin{bmatrix}a_{ij}\end{bmatrix}$ over the set of nonnegative integers such that:
\begin{equation}
a_{ij} = \text{the number of edges connecting } v_i \text{ and } v_j \ \forall i, j = 1,2,...,n.
\end{equation}

# Matrix Representation of an Directed Graphs.
---

- Let $G$ be a directed graph with ordered vertices $v_1 , v_2 , . . . , v_n$ . The adjacency matrix of $G$ is the $n \times n$ $\textbf{A} = \begin{bmatrix}a_{ij}\end{bmatrix}$ over the set of nonnegative integers such that:
\begin{equation}
a_{ij} = \text{the number of arrows from } v_i \text{ to } v_j \forall i, j = 1,2,...,n.
\end{equation}

The adjacency matrix of a directed graph are the counts of all arrows heads points from $v_i$ to $v_j$.

\begin{equation}
\textbf{A} = a_{ij} = 
\begin{bmatrix}
\begin{array}{c|ccccc}
& A & B & C \\
\hline
A & 0 & 1 & 1 \\
B & 0 & 0 & 1 \\
C & 1 & 1 & 0 \\
\end{array}
\end{bmatrix}
\end{equation}

# Characterizing a Tree
---

- Any tree that has more than one vertex has at least one vertex of degree 1.

Proof:

Let $T$ be a particular but arbitrarily chosen tree that has more than one vertex, and consider the following algorithm:
1. Pick a vertex, $v$, of $T$ and let $e$ be an edge incident on $v$. $\textbf{NOTE:}$ If there were no edge incident on $v$, then v would be an isolated vertex. But this would contradict the assumption that T is connected (since it is a tree) and has at least two vertices.

2. While $deg(v) > 1$, do:
   - Choose $e'$ to be an edge incident on $v$ such that $e' \neq e$. $\textbf{NOTE:}$ Such an edge exists because by definition $deg(v) > 1$ and so there are at least two edges incident on v.
   - Let $v'$ be the vertex at the other end of $e'$ from $v$. $\textbf{NOTE:}$ Since $T$ no loop should exist.
   - Let $e = e'$ and $v = v'$.
   - repeat the while loop until such a time where $deg(v) = 1$.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.Graph()

# Add nodes to the graph
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")
G.add_node("G")
G.add_node("H")
G.add_node("I")


# Add directed edges (arrows) - (source, target)
G.add_edge("A", "B")
G.add_edge("A", "C")
G.add_edge("B", "D")
G.add_edge("B", "E")
G.add_edge("B", "F")
G.add_edge("C", "G")
G.add_edge("C", "H")
G.add_edge("C", "I")


### Step 3: Visualize the Graph

# Draw the graph
pos = nx.planar_layout(G)  # positions for all nodes

# nodes
nx.draw_networkx_nodes(G, pos, node_size=700)

# edges
nx.draw_networkx_edges(G, pos, edgelist=G.edges())

# labels
nx.draw_networkx_labels(G, pos, font_size=15, font_family='sans-serif')

plt.axis('off')  # Turn off the axis
plt.show()  # Display the graph
```

# Leaves and Branches
---
- A vertex of degree 1 in $T$ is called a terminal vertex (or a leaf )
- A vertex of degree greater than 1 in $T$ is called an internal vertex (or a branch vertex).
- $\textbf{ALSO NOTE THAT: }$ For any positive integer $n$, any tree, $T$, with $n$ vertices has $n − 1$ edges.

# Rooted Trees
---

- A rooted tree is a tree in which there is one vertex that is distinguished from the others and is called the root.

- The level of a vertex is the number of edges along the unique path between it and the root.

- The height of a rooted tree is the maximum level of any vertex of the tree.

-  Given the root or any internal vertex v of a rooted tree, the $\textbf{children}$ of $\textbf{v}$ are all those vertices that are adjacent to $\textbf{v}$ and are one level farther away from the root than $\textbf{v}$. If $\textbf{w}$ is a child of $\textbf{v}$, then $\textbf{v}$ is called the $\textbf{parent}$ of $\textbf{w}$, and two distinct vertices that are both children of the same parent are called $\textbf{siblings}$. Given two distinct vertices $\textbf{v}$ and $\textbf{w}$, if $\textbf{v}$ lies on the unique path between $\textbf{w}$ and the root, then $\textbf{v}$ is an $\textbf{ancestor}$ of $\textbf{w}$ and $\textbf{w}$ is a $\textbf{descendant}$ of $\textbf{v}$.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a balanced binary tree
T = nx.balanced_tree(2, 4)

# Function to calculate depth of nodes
def get_depth(node, height):
    # Nodes in a balanced binary tree can be found through this formula
    depth = 0
    while node != 0:
        node = (node - 1) // 2
        depth += 1
    return depth

# Label edges with layers
for edge in T.edges():
    # The layer can be considered as the depth of the shallowest node (i.e., closest to root)
    layer = min(get_depth(edge[0], 5), get_depth(edge[1], 5))
    T.edges[edge]['layer'] = f'Lvl {layer + 1}'

plt.figure(figsize=(10, 8))

# Draw the graph with node labels
pos = graphviz_layout(T, prog="dot")  # Set a fixed layout for consistency in visualization
nx.draw(T, pos, with_labels=True, node_color='lightblue')

# Draw edge labels based on layers
edge_labels = nx.get_edge_attributes(T, 'layer')
nx.draw_networkx_edge_labels(T, pos, edge_labels=edge_labels)
plt.title('Balanced Tree with Layered Edges')
plt.show()
```
